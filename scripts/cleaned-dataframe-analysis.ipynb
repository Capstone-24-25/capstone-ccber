```python
def getString(current, current_null):
    text = ""
    text = text + "The sample '" + str(current['database_id']) + "' has been identified as a(n) " + str(current['species']) + " from the family: " + str(current['family'])
    # include more information about what the database is here.

    if current_null['subfamily'] == False:
        text = text + ", subfamily: " + str(current['subfamily']) 

    text = text + ", genus: " + str(current['genus'])

    if current_null['subspecies'] == False:
        text = text + ", subspecies: " + str(current['subspecies'])

    text = text + ", species: " + str(current['species']) + ", specific epithet: " + str(current['specificEpithet'])

    if current_null['acceptedNameUsage'] == False:
        text = text + ", with an accepted name of: " + str(current['acceptedNameUsage'])

    if current_null['taxonRank'] == False:
        text = text + ", and a taxon rank of: " + str(current['taxonRank'])

    text = text + ", whose scientific name for this specific species was authored by " + str(current['scientificNameAuthorship']) + ". "

    if current_null['identificationQualifier'] == False:
        text = text + "The database has included the following identification qualifier for this sample: " + str(current['identificationQualifier']) + ". "

    if current_null['higherClassification'] == False:
        text = text + "This species also has a higher classification of: " + str(current['higherClassification']) + ". "

    if current_null['identificationReferences'] == False:
        text = text + "Identification references for this sample were provided: " + str(current['identificationReferences']) + ". "

    if current_null['typeStatus'] == False: #figure out typeStatus
        text = text + "The current type status for this sample was provided as: " + str(current['typeStatus']) + ". "

    if current_null['verbatimIdentification'] == False: 
        text = text + "The verbatim identification for this sample was: " + str(current['verbatimIdentification']) + ". "

    if (current_null['identifiedBy'] == False) & (current_null['dateIdentified'] == False):
        text = text + "This sample was identified by " + str(current['identifiedBy']) + " on " + str(current['dateIdentified']) + ". "
    elif current_null['identifiedBy'] == False:
        text = text + "This sample was identified by " + str(current['identifiedBy']) + ". "
    elif current_null['dateIdentified'] == False:
        text = text + "This sample was identified on " + str(current['dateIdentified']) + ". "
    
    text = text + "The sample was found at (" + str(current['decimalLatitude'].round(2)) + ", " + str(current['decimalLongitude'].round(2)) + ") rounded to 2 decimal points. Specifically, the bee was collected" 

    if current_null['island'] == False:
        text = text + " on " + str(current['island']) + " island,"
    if current_null['county'] == False:
        text = text + " in " + str(current['county']) + " county,"
    if current_null['municipality'] == False:
        text = text + " in " + str(current['municipality']) + " municipality,"
    if current_null['locality'] == False:
        text = text + " in " + str(current['locality']) + " locality,"
    if current_null['stateProvince'] == False:
        text = text + " in " + str(current['stateProvince']) + " state/province,"
    if current_null['country'] == False:
        text = text + " in " + str(current['country'])
    if current_null['continent'] == False:
        text = text + " in " + str(current['continent'])
    
    text = text + " on " + str(current['eventDate'])
    if current_null['eventTime'] == False:
        text = text + str(current['eventTime'])
    
    text = text + ". "

    if current_null['basisOfRecord'] == False:
        text = text + "The basis of record included for this sample is: " + str(current['basisOfRecord']) + ". "

    if current_null['samplingProtocol'] == False:
        text = text + "The sampling protocol for this sample was: " + str(current['samplingProtocol']) + ". "

    if current_null['catalogNumber'] == False:
        text = text + "The catalog number for this sample is '" + str(current['catalogNumber']) + "'. "

    if current_null['institutionCode'] == False:
        text = text + "The institution code associated with this sample is: '" + str(current['institutionCode']) + "'. "

    if current_null['verbatimScientificName'] == False:
        text = text + "The verbatim scientific name for this sample is: " + str(current['verbatimScientificName']) + ". "

    if current_null['verbatimEventDate'] == False:
        text = text + "The verbatim event date associated with this sample is: " + str(current['verbatimEventDate']) + ". "

    if current_null['associatedTaxa'] == False:
        text = text + "Associated taxa are: " + str(current['associatedTaxa']) + ". "

    if current_null['associatedOrganisms'] == False:
        text = text + "Associated organisms are: " + str(current['associatedOrganisms']) + ". "

    if current_null['fieldNotes'] == False:
        text = text + "Field notes were included with this sample and are as follows: \"" + str(current['fieldNotes']) + "\". "

    if current_null['sex'] == False:
        text = text + "The sex of the sample is " + str(current['sex']) + ". "

    if current_null['associatedReferences'] == False:
        text = text + "Associated references with this sample were are as follows: '" + str(current['associatedReferences']) + "'. " 

    if current_null['bibliographicCitation'] == False:
        text = text + "The bibliographic citation for this sample is: '" + str(current['bibliographicCitation']) + "'. "
    
    if current_null['references'] == False:
        text = text + "References for this sample are as follows: '" + str(current['references']) + "'. "

    text = text + "\n===\n"

    return text
```

```python
null_table = data_full_mincol.isnull()

for i in range(len(data_full_mincol)):
    current = data_full_mincol.iloc[i]
    current_null = null_table.iloc[i]

    with open("../data/cleaned-database.txt", 'a', encoding = 'latin-1') as file:
        file.write(getString(current, current_null))

    if i % 10000 == 0:
        print(i)
```
